```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import matplotlib.pyplot as plt

# Чтение данных из файла (измените на ваш файл)
df = pd.read_csv('dataset.cvs')
# Строим график
colon = ['price','copiesSold', 'revenue', 'avgPlaytime',  'reviewScore']
for i in colon:
    for j in colon:
        if (i == j): continue
        # Масштабирование данных (MinMaxScaler для reviewScore)
        scaler_x = StandardScaler()
        scaler_y = MinMaxScaler()

        df['scaler_x'] = scaler_x.fit_transform(df[[i]])
        df['scaler_y'] = scaler_y.fit_transform(df[[j]])

        scaled_data = df[['scaler_x', 'scaler_y']]

        kmeans = KMeans(n_clusters=3)
        kmeans.fit(scaled_data)
        df['cluster'] = kmeans.labels_

        # Центроиды в исходном масштабе
        centroids_scaled = kmeans.cluster_centers_
        centroids = np.zeros_like(centroids_scaled)
        centroids[:, 0] = scaler_x.inverse_transform(centroids_scaled[:, 0].reshape(-1,1)).flatten()
        centroids[:, 1] = scaler_y.inverse_transform(centroids_scaled[:, 1].reshape(-1,1)).flatten()


        # Визуализация
        plt.figure(figsize=(8, 6))
        plt.scatter(df[i], df[j], c=df['cluster'], cmap='viridis')
        plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, color='red', label='Centroids')
        plt.xlabel(i)
        plt.ylabel(i)
        title = "KMeans :" + j + " и " + i
        plt.title(title)
        plt.legend()
        plt.show()

        print(df.head())
```
